```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
ds=pd.read_csv('/content/penguins_size.csv')
ds.head()
```

```python
ds.info()
```

```python
ds.isnull().sum()
```

```python
ds.dropna(inplace=True)
```

```python
sns.countplot(x='sex',data=ds)
plt.title("Count of male and female penguins")
plt.xlabel('Gender')
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='island',data=ds,palette='rainbow')
plt.title("Count of penguins per island")
plt.xlabel('Island')
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='island',hue='sex',data=ds,palette='rainbow')
plt.title("Count of penguins per island according to sex")
plt.xlabel('Island')
plt.ylabel("Count")
plt.show()
```
